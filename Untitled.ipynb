```python
def read_file(filename):
    with open(filename) as f:
        text = f.read()
    return text
```

```python
file_list = [i for i in listdir() if '.txt' in i]
sample = list()
qt_files = len(file_list)
for i in range(300):
    sample.append(file_list[randint(0, qt_files)])
```

```python
contents = list()
for i in sample:
    try:
        contents.append(read_file(i))
    except UnicodeDecodeError as e:
        print(e)
```

```python
im = list()
for i in file_list:
    try:
        with open(i) as f:
            content = [i for i in f.readlines() if i]
    except UnicodeDecodeError as e:
        print(e)
        continue

    for j in content[7:]:
        if ':' not in j:
            im.append(j)
            break
```

```python
data = [i.replace('\n', '').replace(' ', '') for i in im if '\xa0' not in i]
```

```python
dog = set(data)
```

```python
dog_data = [(i,data.count(i)) for i in dog]
```
