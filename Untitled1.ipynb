```python
import re
from sys import argv
from bs4 import BeautifulSoup
from os import listdir

def get_text(file_name):
    with open(file_name) as f:
        t = f.read()
    bs = BeautifulSoup(t, "html5lib")
    text = bs.text.replace('E M E N T A', 'EMENTA')
    text = text.replace('D E S P A C H O', 'DESPACHO')
    text = text.replace('D E C I S Ã O', 'DECISÃO')
    text = text.replace('A C Ó R D Ã O', 'ACORDÃO')
    return text
```

```python
files = listdir('data/sample/')
```

```python
for i in files[:50]:
    text = get_text(i)
    print(re.findall('\n(EMENTA|DECISÃO|DESPACHO|ACÓRDÃO).*\n', text))
    isthere = re.compile('(EMENTA|DECISÃO|DESPACHO|ACÓRDÃO)\n.+(?=(DECISÃO|ACÓRDÃO))?', flags=re.DOTALL)
    print(isthere.search(text).group())
```
